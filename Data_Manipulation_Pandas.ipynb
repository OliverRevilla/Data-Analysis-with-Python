# Chapter 2: Aggregating Data

```python
# Now it`s time to analyze "sales", a date subset with Waltmart's sales.
df2 = pd.read_csv('sales_subset.csv', sep = ',')
df2.head()
```

## Summary statistics

```python
df2.info()
```

```python
df2['weekly_sales'].mean()
```

```python
df2['weekly_sales'].median()
```

```python
df2['date'].max()
```

```python
df2['date'].min()
```

```python
# A custom IQR function
def rango_intercuartil(column):
    return column.quantile(0.75) - column.quantile(0.25)

# Function agg
df2['weekly_sales'].agg(rango_intercuartil)
```

```python
df2[['weekly_sales','fuel_price_usd_per_l','temperature_c']].agg([rango_intercuartil, np.median])
```

```python
df2['weekly_sales'].cumsum()
```

```python
df2['weekly_sales'].cummax()
```

## Counting

#### Dropping duplicates

```python
df2.drop_duplicates(subset = ['store','type'])
```

```python
df2[df2['is_holiday']].drop_duplicates(subset = ['date'])
```

```python
df2['type'].value_counts()
```

```python
df2['type'].value_counts(normalize = True)
```

```python
df2['date'].value_counts()
```

## Grouped summary statistics

```python
# Group by type
df2.groupby('type')['weekly_sales'].sum()
```

```python
df2.groupby(['type','is_holiday'])['weekly_sales'].count()
```

```python
df2.groupby('type')['weekly_sales'].agg([np.min,np.max,np.mean,np.median])
```

```python
df2.groupby('type')['unemployment','fuel_price_usd_per_l'].agg([np.min,np.max,np.mean,np.median])
```

Pivot Tables

Pivot tables are the standard way of aggregating data in spreadsheets. 

In pandas, pivot tables are essentially just another way of performing grouped calculations. 

That is, the .pivot_table() method is just an alternative to .groupby().

```python
df2.pivot_table(values = 'weekly_sales', index = 'type')
```

```python
df2.pivot_table(values = 'weekly_sales',index = 'type', aggfunc = ['mean','median'])
```

```python
df2.pivot_table(values = 'weekly_sales', index = 'type', columns = 'is_holiday')
```

```python
df2.pivot_table(values = 'weekly_sales', index = 'department', columns = 'type', fill_value = 0, margins = True)
```

# Chapter 3: Slicing and Indexing Data

```python
df3 = pd.read_csv('temperatures.csv', sep = ',')
df3.head()
```

```python
df3.set_index('city')
```

```python
df3.reset_index()
```

```python
df3.reset_index(drop = True)
```

```python
cities = ['Moscow','Saint Petersburg']
df3[df3['city'].isin(cities)]
```

```python
df3.set_index('city').loc[cities] # It`s important ta have set the index previously
```

```python
rows_to_keep = [('Brazil','Rio De Janeiro'),('Pakistan','Lahore')]
df3.set_index(['country','city']).loc[rows_to_keep]
```

```python
df3.set_index(['country','city']).sort_index()
```

```python
df3.set_index(['country','city']).sort_index(level = 'city')
```

```python
df3.set_index(['country','city']).sort_index(level = ['country','city'], ascending = [True,False])
```
