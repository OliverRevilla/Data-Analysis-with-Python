## Relational plots and subplots

```python
# relplot(): 
sns.scatterplot(x = "total_bill", y = "tip", data = tips)
plt.show()
```

```python
sns.relplot(x = "total_bill", y = "tip",\
            data = tips, kind = "scatter", col = "smoker")
plt.show()
```

```python
sns.relplot(x = "total_bill", y = "tip",\
            data = tips, kind = "scatter",\
            col = "smoker", row = "time")
plt.show()
```

```python
sns.relplot(x = "total_bill", y = "tip",\
            data = tips, kind = "scatter",\
            col = "day",\
            col_wrap = 3, # This parameter specify the number of graphs x row,
            col_order = ["Thur","Fri","Sat","Sun"])
plt.show()
```

```python
sns.relplot(x = "total_bill", y = "tip",\
            data = tips,
            kind = "scatter",
            size = "size")
plt.show()
```

```python
sns.relplot(x = "total_bill", y = "tip",\
            data = tips,
            kind = "scatter",
            size = "size",
            hue = "size")
plt.show()
```

```python
sns.relplot(x = "total_bill",
            y = "tip",
            data = tips,
            kind = "scatter",
            hue = "smoker",
            style = "smoker")
plt.show()
```

```python
# Alpha: transparency
sns.relplot(x = "total_bill",
            y = "tip",
            data = tips,
            kind = "scatter",
            alpha = 0.4)
plt.show()
```

## Box plots

```python
sns.catplot(x = "time",
            y = "total_bill",
            data = tips,
            kind = "box")
plt.show()
```

## Adding titles and labels

    Object type:    Plot Types
    FacetGrid       relplot(), catplot()
    AxesSubplot     scatterplot(), countplot()...

```python
gdp = pd.read_csv("WorldBank_GDP.csv")
gdp.head()
```

```python
g = sns.catplot(x = "Year",
            y = "GDP",
            data = gdp,
            kind = "box")
g.fig.suptitle("New Title", y = 1) # By default y = 1.0
plt.show()
```

     FacetGrid               AxesSubplot
     g = sns.catplot(...)    g = sns.boxplot(...)
     g.fig.suptitle("...")   g.set_title("...")
 

```python
sns.set_style("whitegrid")
g = sns.catplot(x = "Year",
                y = "GDP",
                data = gdp,
                kind = "bar",
                col = "Country Name")
# Adding general title
g.fig.suptitle("New title", y = 1.10)
# Adding title to each grid
g.set_titles("This is {col_name}")
# Adding axis labels
g.set(xlabel = "New X Label",
      ylabel = "New Y Label")
# Rotating x-axis tick labels
plt.xticks(rotation = 90)
plt.show()
```
